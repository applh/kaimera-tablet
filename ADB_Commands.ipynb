# Kaimera Tablet - ADB Utility
Use this notebook to quickly execute common ADB commands for the Kaimera Tablet project.

### Build Debug APK (via Skills)
Runs the project's build debug script.

```python
!chmod +x .agent/skills/AndroidDevelopment/scripts/build_debug.sh
!.agent/skills/AndroidDevelopment/scripts/build_debug.sh
```

### List Connected Devices

```python
!adb devices
```

### Install Debug APK
Installs the debug APK generated by Gradle.

```python
!adb install -r app/build/outputs/apk/debug/app-debug.apk
```

### Uninstall App
Removes the Kaimera Tablet app from the device.

```python
!adb uninstall com.kaimera.tablet
```

### Clear App Data
Resets the app to its initial state (clears local storage/databases).

```python
!adb shell pm clear com.kaimera.tablet
```

### Take Screenshot
Captures a screenshot and saves it to the project directory.

```python
!adb exec-out screencap -p > screenshot_$(date +%Y%m%d_%H%M%S).png
print("Screenshot saved.")
```

### View Logcat (Last 100 lines)
Shows the recent logs for the Kaimera Tablet package.

```python
!adb logcat -d -t 100 | grep "com.kaimera.tablet"
```

### Reboot Device

```python
!adb reboot
```

### Save Full Bug Report
Dumps the entire logcat buffer to a file named `bug_report.log`.

```python
!adb logcat -d > bug_report.log
```
